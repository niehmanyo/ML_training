<b>Uniform distributions are probability distributions with equally likely outcomes.</b> 
In a discrete uniform distribution, outcomes are discrete and have the same probability. 
In a continuous uniform distribution, outcomes are continuous and infinite. 
In a normal distribution, data around the mean occur more frequently.

```python
import numpy as np
x = np.random.uniform(0.0,5.0,250) 
# create an array containing 250 random floats between 0 and 5

print(x)
```

## Histogram (Áõ¥ÊñπÂõæüìä)

To visualize the data we collected.
This is a function in *Python module Matplotlib*

```python
import matplotlib.pyplot as plt
import numpy as np 

x = np.random.uniform(1.0,5.0,50)
plt.hist(x,10)
plt.show()
print(x)
```
